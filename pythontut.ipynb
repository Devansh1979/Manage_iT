```python
df=pd.DataFrame({'name':['Alice','Julie','Alice'],
                 'toy':['Ship','plane',np.nan],
                 'born':[pd.NaT,pd.Timestamp('1940-12-25'),pd.NaT]})
df.dropna(how='all',axis=1)
```
